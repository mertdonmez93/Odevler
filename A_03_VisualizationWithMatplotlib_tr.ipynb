In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False,index_col=['date'])
df.sort_values(by='date',inplace=True)
```

```python
spain_data = df[df['location'].isin(['Spain'])]
spain100 = spain_data.iloc[-100:]

france_data = df[df['location'].isin(['France'])]
france100 = france_data.iloc[-100:]

germany_data = df[df['location'].isin(['Germany'])]
germany100 = germany_data.iloc[-100:]

italy_data = df[df['location'].isin(['Italy'])]
italy100 =italy_data.iloc[-100:]
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
plt.title('Cases for Countries', 
          fontstyle = 'normal', # {'normal', 'italic', 'oblique'}
          fontsize = 20,
          fontweight = 'normal', # {a numeric value in range 0-1000, 'normal', 'regular', etc.}
          color = "darkred",
          fontfamily = 'Century Gothic'
         )
plt.plot(spain100['total_cases'], c = "r", ms=10, ls="--", lw = 3, label = 'Spain')
plt.plot(france100['total_cases'], c = "g",ms=10, ls=":" ,lw = 3, label = 'France')
plt.plot(germany100['total_cases'], c = "b",ms=10, ls="-", lw = 3, label = 'Germany')
plt.plot(italy100['total_cases'], c = "y",ms=10, ls="--", lw = 3, label = 'Italy')
plt.xticks(rotation = 90, fontsize = 9, c = "r")
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend(title = 'Countries', loc = 4)
plt.subplot(1,2,2)
plt.title('Deaths for Countries', 
          fontstyle = 'normal', # {'normal', 'italic', 'oblique'}
          fontsize = 20,
          fontweight = 'normal', # {a numeric value in range 0-1000, 'normal', 'regular', etc.}
          color = "darkred",
          fontfamily = 'Century Gothic'
         )
plt.plot(spain100['total_deaths'], c = "r", lw = 6, label = 'Spain')
plt.plot(france100['total_deaths'], c = "g", lw = 6, label = 'France')
plt.plot(germany100['total_deaths'], c = "b", lw = 6, label = 'Germany')
plt.plot(italy100['total_deaths'], c = "y", lw = 6, label = 'Italy')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.xticks(rotation = 90, fontsize = 9, c= "r")
plt.legend(title = 'Countries', loc = 4)
plt.show()
```
