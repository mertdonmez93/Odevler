**(1)** Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False,index_col=['location'])
df.sort_values(by='date',inplace=True)
```

```python
newdf = df[['date','total_cases']]
df_filtered=newdf[newdf['date']=='2021-03-26']
ddf=df_filtered.sort_values(by='total_cases' ,ascending=True,na_position='first')

plt.figure(figsize=(14, 5))
plt.title('Total Cases')
plt.bar(ddf.iloc[176:196].index,ddf.iloc[176:196]['total_cases'])
plt.xlabel('Countries')
plt.ylabel('Total Cases on 2021-03-26')
plt.xticks(rotation = 90, fontsize = 9)
plt.show()
```

**(2)** Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
df_brazil = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False,index_col=0)
ddf_brazil = df_brazil[(df_brazil['location']=='Brazil')]

plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
X = ddf_brazil['new_cases']
plt.hist(X)

plt.subplot(1,3,2)
X = ddf_brazil['new_cases']
plt.hist(X, bins=50)

plt.subplot(1,3,3)
X = ddf_brazil['new_cases']
plt.hist(X, bins=100)
plt.show()

plt.show()
```

**(3)** Plot a scatter plot of new cases and new death for Germany and France.

```python
dfnew_cases = df_brazil[df_brazil['location'].isin(['Germany', 'France'])]
plt.figure(figsize=(10,10))
plt.title('New Cases & New Deaths')
for numbers in ['Germany','France']:
    plt.scatter(dfnew_cases[dfnew_cases.location == numbers].new_cases,
               dfnew_cases[dfnew_cases.location == numbers].new_deaths,label = numbers)
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.legend(title='Countries',loc=4)
plt.show()
```

**(4)** Plot a boxplot for daily deaths for any country you choose.

```python
ddf_brazil = df_brazil[(df_brazil['location']=='Brazil')]
plt.figure(figsize=(10,4))
plt.title('Daily Deaths in Brazil')
plt.boxplot(ddf_brazil['new_cases'].dropna())
plt.show()
```

**(5)** Calculate the total case for each continent and plot a pie chart 

```python
df.sort_values(by='total_cases',inplace=True)
dfcontinents=df[(df['location'].isin(['Asia', 'Africa', 'Europe', 'North America', 'South America']))&(df['continent'].isna() )& (df['date']=='2021-03-26')]
continents = ['Asia', 'Africa', 'Europe', 'North America', 'South America']
plt.pie(dfcontinents['total_cases'],labels= continents, autopct='%1.2f%%', shadow=True, startangle=0)
plt.show()
```
