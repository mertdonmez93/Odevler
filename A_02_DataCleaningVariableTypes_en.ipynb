**(3)** Is it possible to convert categorical variable into numerical one? If so, how can you make that?

```python
#By changing the types of the variables.
# 0 for black and white
# 1 for color
def make_discrete(color):
    if color == "Color":
        return 1
    elif color == " Black and White":
        return 0
    else:
        return "NaN"

movie['color_group'] = movie["color"].apply(make_discrete)
display(movie.tail(15))
```
