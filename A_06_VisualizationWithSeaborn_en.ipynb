In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Seaborn library. 

**(1)** Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
dftotal_deaths = df[(df.location.isin(["Spain", "France","Germany", "Italy"])) & (df['date']>='2020-04-01')& (df['date']<='2020-08-01')]
#ödevde grafik x axis aralığı 2020-04-01 ve 2020-08-01 tarihleri arasında çizilmiştir.

plt.figure(figsize=(10, 4), dpi = 100)
sns.lineplot(x='date',y='total_deaths',hue= "location" ,data = dftotal_deaths)

plt.xticks(rotation = 45, fontsize = 9)
plt.title('Total Deaths', c = "r")
plt.xlabel('Dates', c="b")
plt.ylabel('Count of Dates', c="b")

plt.show()
```

**(2)** Plot a bar plot with seaborn for average death number that compares continents.

```python
dfcontinent_deaths = df
plt.figure(figsize=(9, 6), dpi = 100)
sns.barplot(x='continent',y = 'total_deaths', data = dfcontinent_deaths)
plt.xlabel('Continents', c="b")
plt.show()
```

**(3)**  Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
df_q4 = df[(df.location.isin(["Turkey"]))]
plt.figure(figsize=(10,16),dpi=600)

plt.subplot(2,2,1)
sns.distplot(df_q4.new_deaths.dropna(),kde=False,bins=50)
plt.title("New Deaths \n(Bins:50 / KDE = False)")
plt.ylabel('Count')
plt.xlabel('New Deaths')

plt.subplot(2,2,2)
sns.distplot(df_q4.new_deaths.dropna(),kde=False,bins=20)
plt.title("New Deaths \n(Bins:20 / KDE = False)")
plt.ylabel('Count')
plt.xlabel('New Deaths')

plt.subplot(2,2,3)
sns.distplot(df_q4.new_deaths.dropna(),kde=True,bins=50)
plt.title("New Deaths \n(Bins:50 / KDE = True)")
plt.ylabel('Normalized Count')
plt.xlabel('New Deaths')

plt.subplot(2,2,4)
sns.distplot(df_q4.new_deaths.dropna(),kde=True,bins=20)
plt.title("New Deaths \n(Bins:20 / KDE = True)")
plt.ylabel('Normalized Count')
plt.xlabel('New Deaths')

plt.show()
```

**(4)** Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
df_q4 = df[df.location.isin(["France","Germany"])]
plt.figure(figsize=(20,8),dpi=600)

plt.subplot(1,3,1)
sns.boxplot(x = 'location',  y = 'new_deaths', data = df_q4, palette = ['#FF914D', '#22B573'])
plt.title("Box Plot")
plt.xlabel("Country")
plt.ylabel("New Deaths")

plt.subplot(1,3,2)
sns.violinplot(x = 'location',  y = 'new_deaths', data = df_q4, palette = ['#FF914D', '#22B573'])
plt.title("Violin Plot")
plt.xlabel("Country")
plt.ylabel("New Deaths")


plt.subplot(1,3,3)
sns.swarmplot(x = 'location',  y = 'new_deaths', data = df_q4, palette = ['#FF914D', '#22B573'])
plt.title("Box Plot")
plt.xlabel("Country")
plt.ylabel("New Deaths")


plt.show()
```
