```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

- Like in the previous lesson, build a linear regression model where your target variable is the difference between the *apparenttemperature* and the *temperature*. As explanatory variables, use *humidity* and *windspeed*. Now, estimate your model using OLS. What are the R-squared and adjusted R-squared values? Do you think they are satisfactory? Why? 

```python
weather_df.insert(3, "Temperature Difference", 
               (weather_df["Temperature (C)"] - weather_df["Apparent Temperature (C)"]))
```

```python
X = weather_df[["Humidity", "Wind Speed (km/h)"]]
y = weather_df["Temperature Difference"]

X = sm.add_constant(X)
results = sm.OLS(y, X).fit()
results.summary()
```

**R-squared = Adj.R-squared = 0.288 Which is very low. In brief, our model %28 of the variance for a dependent variable that's explained by an independent variable or variables.**

- Next, include the interaction of *humidity* and *windspeed* to the model above and estimate the model using OLS. Now, what is the R-squared of this model? Does this model improve upon the previous one? 


```python
weather_df.insert(5, "Humidity*Wind Speed", 
               (weather_df.Humidity * weather_df["Wind Speed (km/h)"]))

weather_df["Humidity x Wind Speed"] = weather_df.Humidity * weather_df["Wind Speed (km/h)"]
XX = weather_df[["Humidity", "Wind Speed (km/h)","Humidity*Wind Speed"]]
y = weather_df["Temperature Difference"]

XX = sm.add_constant(XX)
results_new = sm.OLS(y, XX).fit()
results_new.summary()
```

**R-squared increased to 0.341. Results can explain %34 of the variance for a temperature difference. Which is still low.**
