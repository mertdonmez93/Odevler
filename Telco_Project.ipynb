```python
def make_ordinal(Contract):
    if Contract == "Month-to-month":
        return "Pay as go"
    else: 
        return "Has plan"
    
df['HasContract'] = df["Contract"].apply(make_ordinal)
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.boxplot(x = 'Churn',  y = 'MonthlyCharges',hue = 'Contract', data = df)
plt.title("Contract vs. Monthly Charges")

plt.subplot(1,2,2)
sns.boxplot(x = 'Churn',  y = 'Total_Charges',hue = 'HasContract', data = df)
plt.title("Internet Service vs. Total Charges")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.boxplot(x = 'Churn',  y = 'MonthlyCharges',hue = 'HasContract', data = df)
plt.title("Contract vs. Monthly Charges")
plt.subplot(1,2,2)
sns.boxplot(x = 'Churn',  y = 'Total_Charges',hue = 'HasContract', data = df)
plt.title("Contract vs. Total Charges")
plt.show()
```

```python
import scipy.stats as stats
from scipy.stats.mstats import winsorize
import warnings
warnings.filterwarnings('ignore')

df["Winsorized_Total_Charges"] = winsorize(df['Total_Charges'], (0,0.15))

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.boxplot(x = 'Churn',  y = 'Winsorized_Total_Charges',hue = 'HasContract', data = df)
plt.title('Average Number of Winsorize Total Charges')

plt.subplot(1,2,2)
sns.barplot(df['Churn'],df["Winsorized_Total_Charges"])
plt.title('Average Number of Winsorize Total Charges')
plt.show()
```

**Average Total Charge for the customers not churned is 1.5 times higher than customers  churned**

```python
stats.ttest_ind([df["Churn"]=="Yes"],df["Total_Charges"],axis=0,equal_var=True)
```

```python
def make_ordinal(SeniorCitizen):
    if SeniorCitizen == 1:
        return "SeniorCitizen"
    else: 
        return "Not SeniorCitizen"
    
df['Senior_Citizen'] = df["SeniorCitizen"].apply(make_ordinal)
```

```python
import math
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
df= pd.concat([df, pd.get_dummies(df["PaymentMethod"])], axis=1)
df= pd.concat([df, pd.get_dummies(df["InternetService"])], axis=1)
df= pd.concat([df, pd.get_dummies(df["HasContract"])], axis=1)
df= df.rename(columns={'No': 'NoInternet'})
df= pd.concat([df,pd.get_dummies(df["Senior_Citizen"],drop_first=True)],axis=1)
df= pd.concat([df,pd.get_dummies(df["Churn"],drop_first=True)],axis=1)
df= df.rename(columns={'Yes': 'Churn'})
df.head()
```
